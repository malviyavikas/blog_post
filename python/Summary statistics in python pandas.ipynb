## Summary Statistics in Python Pandas
describe() funcion is used to get the summary statistics in pandas dataframe. describe() gives mean,std,IOR values.

. By default describe() excludes the character columns and gives summary statistics of numeric columns.<br>
. To get the statistics of both the numercial and character coumn we need to add variable names include='all'

### Lets Create the dataframe

```python
import pandas as pd
import numpy as np
import random

#Create a dataframe
raw_data = {'first_name': ['Sam','Ziva','Kia','Robin','Kim','Ram','Zen','John','Ariel','Ron'], 
         'degree': ['PhD','MBA','PhD','MS','MBA','PhD','PhD','MS','PhD','PhD'],
         'nationality': ['USA','India','India','USA','India','UK','USA','USA','India','UK'],
         'age': [25, 29, 19, 21, 33, 38, 24, 26, 32, 29]}
df = pd.DataFrame(raw_data)

#Randomly generate Salary between 80K to 150K having steps (multiple) of 10K
df['salary']=0
df['salary']=df['salary'].apply(lambda x: random.randrange(80000,150000,10000))
 
df
```

### Pandas- Summary Statistic of the numeric columns:

describe() Function gives the mean, std and IQR values. It excludes character column and calculate summary statistics only for numeric columns.<br><br>
so the output will be

```python
df.describe()
```

### Pandas – Summary Statistic of the character columns:

describe() Function with include=’all’ gives the summary statistics of all the columns.

So the output will be

```python
df.describe(include='all')
```
