## mode() Function in Python statistics

Python is very popular language when it comes to statistics and analysis.python 3 provides statistics module which has important function mean(),median() and mode() etc. In this tutorial we will learn about the mode() function it returns mean of the dataset.


### Statistics Mode

The mode is the statistical term that refers to the most frequently occurring number found in a set of numbers. The mode is detected by collecting and organizing data to count the frequency of each result.

### Mode Function in Python

```python
# calculate arithmetic mean
import statistics

# list of positive integer numbers
data1 = [1, 3, 2,4, 5, 7, 9, 2]
  
x = statistics.mode(data1)
  
# Printing the mean
print("Mode is :", x)
```

### Mode of a dataframe in pandas python

```python
#Create a dataframe

import pandas as pd
import numpy as np

raw_data = {'first_name': ['Sam','Ziva','Kia','Robin','Kim','Ram','Zen','John','Ariel','Ron'], 
         'degree': ['PhD','MBA','PhD','MS','MBA','PhD','PhD','MS','PhD','PhD'],
         'age': [25, 29, 19, 21, 33, 38, 24, 26, 32, 29],
         'score': [250, 290, 190, 210, 330, 380, 240, 260, 320, 290]}
df = pd.DataFrame(raw_data)
df
```

```python
# Mean of the dataframe
df.mode()
```

### Column mode of the dataframe

axis=0 argument calculates the column wise mean of the dataframe so the result will be


```python
# column mean of the dataframe
df.mode(axis=0)
```

### Row Mean of the dataframe

axis=1 argument calculates the row wise mean of the dataframe so the result will be

```python
# Row mean of the dataframe
df.mode(axis=1)
```
